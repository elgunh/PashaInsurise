# Loading dataset

```python
df = pd.read_excel("Hackhaton_data.xlsx")
df = df.drop(columns=["ID", "TIN"])
```

```python
df['SALES_TYPE'].fillna("NOT_SOLD", inplace=True)
df['SALES_TYPE'].replace("DIRECT", "SOLD", inplace=True)
df['SALES_TYPE'].replace("ONLINE", "SOLD", inplace=True)
```

```python
df['SALES_TYPE'].unique()
```

# Preprocessing

```python
from sklearn.preprocessing import OrdinalEncoder

columns_to_encode = ['AGE', 'SALARY', 'FK', 'PRD', 'KAT', 'GENDER', 'COMPANY_CLASS',
                     'DIFFICULTY_PERCENTAGE', 'COMPANY_SECTOR', 'CREDIT',
                     'SALES_TYPE', 'SALES', 'HYS', 'CPA', 'MLI', 'YUZ QAT', 'GHS']

ordinal_encoder = OrdinalEncoder()
df[columns_to_encode] = ordinal_encoder.fit_transform(df[columns_to_encode])
```

```python
df
```

# Train-test split

```python
from sklearn.model_selection import train_test_split
x = df.drop('SALES_TYPE', axis=1)
y = df['SALES_TYPE']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, shuffle=True)
```

# Training

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(x_train)

y_pred = kmeans.predict(x_test)

f1, accuracy = f1_score(y_test, y_pred), accuracy_score(y_test, y_pred)
print(f1, accuracy)
```

```python
from sklearn.svm import LinearSVC
linear_svc = LinearSVC(random_state=42)
linear_svc.fit(x_train, y_train)

y_pred = linear_svc.predict(x_test)
f1, accuracy = f1_score(y_test, y_pred), accuracy_score(y_test, y_pred)
print(f1, accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=42)
rfc.fit(x_train, y_train)
y_pred = rfc.predict(x_test)
y_pred = rfc.predict(x_test)
f1, accuracy = f1_score(y_test, y_pred), accuracy_score(y_test, y_pred)
print(f1, accuracy)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating the meshgrid for x and y
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Define the functions
def f1(x, y):
    return x**2 + 2*y**2

def f2(x, y):
    return x**2 - y**2

def f3(x, y):
    return np.sin(np.sqrt(x**2 + y**2))

def f4(x, y):
    return x**2 - y**2 + 0.5*np.sin(np.sqrt(x**2 + y**2))

def f5(x, y):
    return np.sin(x) * np.cos(y)

def f6(x, y):
    return np.cos(np.sqrt(x**2 + y**2))

# Creating a 2x3 subplot grid for 3D plots
fig = plt.figure(figsize=(18, 12))

functions = [f1, f2, f3, f4, f5, f6]
titles = ['HYS', 'CPA', 'MLI',
          'YUZ QAT', 'GHS', 'YUZ QAT']

for i, (function, title) in enumerate(zip(functions, titles), start=1):
    ax = fig.add_subplot(2, 3, i, projection='3d')
    z = function(x, y)
    ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')
    ax.set_title(title)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_zlim(-5, 5)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def mandelbrot(c, max_iter):
    z = c
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = z*z + c
    return max_iter

# Image size (pixels)
width, height = (600, 400)

# Plot window
real = np.linspace(-2, 1, width)
imag = np.linspace(-1, 1, height)

# Create a white canvas
mandelbrot_image = np.zeros((height, width))

# Generate Mandelbrot image
for i, re in enumerate(real):
    for j, im in enumerate(imag):
        c = complex(re, im)
        mandelbrot_image[j, i] = mandelbrot(c, 100)

# Display the fractal
plt.figure(figsize=(10, 7))
plt.imshow(mandelbrot_image, cmap='hot', extent=(-2, 1, -1, 1))
plt.colorbar()
plt.title('WHAT INSURANCE MODELS SHOULD NOT LOOK LIKE:')
plt.xlabel('Re')
plt.ylabel('Im')
plt.show()
```
